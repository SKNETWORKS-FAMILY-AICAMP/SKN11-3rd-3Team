```python
import json
import faiss
import numpy as np
import re
from sentence_transformers import SentenceTransformer

# ë°ì´í„° ë¡œë“œ
with open("game.json", "r", encoding="utf-8") as f:
    data = json.load(f)

texts = [item["text"] for item in data]
game_names = [item["game_name"] for item in data]

# ì„ë² ë”© ëª¨ë¸ ë¡œë“œ ë° ì„ë² ë”© ìƒì„±
model = SentenceTransformer("BAAI/bge-m3")
embeddings = model.encode(texts, normalize_embeddings=True)

# FAISS ì¸ë±ìŠ¤ ìƒì„±
dimension = embeddings.shape[1]
index = faiss.IndexFlatIP(dimension)
index.add(np.array(embeddings))

# ë¹„ìŠ·í•œ ê²Œì„ ì¶”ì²œ í•¨ìˆ˜
def search_similar_games(query, default_k=5):
    number_match = re.search(r'(\d+)\s*ê°œ', query) # ì‚¬ìš©ì ì§ˆë¬¸ì— ë”°ë¥¸ ì •ê·œí‘œí˜„ì‹
    top_k = int(number_match.group(1)) if number_match else default_k # ìˆ«ìê°€ ìˆìœ¼ë©´ tok

    query_embedding = model.encode([query], normalize_embeddings=True)
    D, I = index.search(np.array(query_embedding), k=top_k) # d ìœ ì‚¬ë„ ì ìˆ˜ ë°°ì—´ i í•´ë‹¹í•˜ëŠ” ê²Œì„ì˜ ì¸ë±ìŠ¤ë°°ì—´

    print(f"\nğŸ” ìœ ì‚¬í•œ ê²Œì„ ì¶”ì²œ ({query}):")
    for rank, (idx, score) in enumerate(zip(I[0], D[0]), start=1):
        print(f"{rank}. {game_names[idx]} (ìœ ì‚¬ë„: {score:.4f})")

# ì‚¬ìš©ì ì…ë ¥
user_query = input("ğŸ® ì›í•˜ì‹œëŠ” ê²Œì„ì´ ìˆë‚˜ìš”? :")
search_similar_games(user_query)
```
